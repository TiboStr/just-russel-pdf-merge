```python
import pandas as pd
import requests
import os
import subprocess
import sys
```

```python
project_id = "just-russel"
query = "SELECT name, image  from dev_data_ingestion.raw_pg_product where image is not NULL"
```

```python
products = pd.read_gbq(query, project_id = project_id, dialect = 'standard')
```

```python
image_links = list((products[products["image"] != ""])["image"])
image_links
```

```python
len(image_links)
```

```python
URL = "https://storage.googleapis.com/just-russel-public-production/uploads/"
PRODUCTS = image_links  # List of image file names or paths

save_dir = "C:/Users/dewae/Downloads"  # change path
os.makedirs(save_dir, exist_ok=True)

if __name__ == "__main__":
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-r", "requirements.txt"])

    for product in PRODUCTS:
        try:
            # Fetch the webp image
            url = URL + product 
            response = requests.get(url)
            response.raise_for_status()  # Ensure the request was successful

            # Save the raw content
            raw_image_path = os.path.join(save_dir, os.path.basename(product))
            with open(raw_image_path, 'wb') as f:
                f.write(response.content)
            print(f"Downloaded and saved {raw_image_path}")

        except requests.exceptions.RequestException as req_err:
            print(f"Network-related error occurred: {req_err}")
        except Exception as e:
            print(f"Failed to download {url}: {e}")
```
